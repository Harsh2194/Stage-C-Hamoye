```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/00471/Data_for_UCI_named.csv')
data.head(5)
```

```python
x = data.drop(columns= ['stab','stabf'] )
y = data[ 'stabf' ]
```

Splitting train and test data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2 , random_state= 1 )
y_train.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state= 1 )
x_train_balanced, y_balanced = smote.fit_resample(x_train, y_train)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
normalised_train_data = scaler.fit_transform(x_train_balanced)
normalised_train_data = pd.DataFrame(normalised_train_data,
columns=x_train_balanced.columns)

normalised_test_data = scaler.fit_transform(x_test)
normalised_test_data = pd.DataFrame(normalised_test_data,
columns=x_test.columns)
```

```python
x_train.shape
```

```python
y_train.shape
```

# Random Forest 

```python
#RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(normalised_train_data, y_balanced)

y_pred = rfc.predict(normalised_test_data)

accuracy = rfc.score(normalised_test_data, y_test)
accuracy
```

# Extra Trees Classifier

```python
#ExtraTreeClassifier
from sklearn.ensemble import ExtraTreesClassifier
etf = ExtraTreesClassifier()
  
# Training the model
etf.fit(normalised_train_data, y_balanced)

y_pred = etf.predict(normalised_test_data)

accuracy = etf.score(normalised_test_data, y_test)
accuracy
```

# Extreme Gradient Boosting Model

```python
#XGBOOST
import xgboost as xgb
xgbr = xgb.XGBClassifier()
xgbr.fit(normalised_train_data, y_balanced)
   
# Predicting the Test set results
y_pred = xgbr.predict(normalised_test_data)
   
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy = xgbr.score(normalised_test_data, y_test)
accuracy
```

# Light Gradient Boosting Mode

```python
#LIGHTGBM
import lightgbm as lgb
from lightgbm import LGBMClassifier

lgbm = LGBMClassifier()
lgbm.fit(normalised_train_data, y_balanced)
 
# Predicting the Target variable
pred = lgbm.predict(normalised_test_data)
print(pred)
accuracy = lgbm.score(normalised_test_data, y_test)
accuracy
```

```python
import matplotlib.pyplot as plt
from sklearn.ensemble import ExtraTreesClassifier
extra_tree_forest = ExtraTreesClassifier(n_estimators = 5,
                                        criterion ='entropy', max_features = 2)
  
# Training the model
extra_tree_forest.fit(x_train, y_train)
  
# Computing the importance of each feature
feature_importance = extra_tree_forest.feature_importances_
  
# Normalizing the individual importances
feature_importance_normalized = np.std([tree.feature_importances_ for tree in 
                                        extra_tree_forest.estimators_],
                                        axis = 0)
```

```python
plt.bar(x_train.columns, feature_importance_normalized)
plt.xlabel('Feature Labels')
plt.ylabel('Feature Importances')
plt.title('Comparison of different Feature Importances')
plt.show()
```
